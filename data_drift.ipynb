```python
import pandas as pd
from evidently.dashboard import Dashboard
from evidently.dashboard.tabs import DataDriftTab
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv("/Users/prudhvisajja/Documents/Data Science MSc/COP528 AI and ML/CourseWork/BankChurners.csv")
```

```python
train_df,test_df = train_test_split(data_df,test_size=0.2,random_state=24)
```

```python
data_drift_dashboard = Dashboard(tabs=[DataDriftTab()])
```

```python
data_drift_dashboard.calculate(reference_data=train_df, current_data=test_df)
```

```python
data_drift_dashboard.show()
```

```python
data_drift_dashboard.save('data_drift.html')
```
